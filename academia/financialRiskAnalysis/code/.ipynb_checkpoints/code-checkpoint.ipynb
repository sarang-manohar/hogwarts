```python
def percent_null(df):
    l = df.shape[0]
    percent_val = []
    for col in df.columns:
        percent_val.append(df[col].isnull().sum()/l)
    df_percent_null = pd.DataFrame({'col':df.columns,
                                    'percent_val':percent_val})
    return df_percent_null
```

```python
eda_data =  percent_null(raw_data).sort_values(by=['percent_val'], ascending=False)
```

```python
from scipy.stats import shapiro as shapiro
stats_l = []
p_val_l = []

for feature in raw_data.columns:
    stats, p_val = shapiro(raw_data[feature])
    stats_l.append(stats)
    p_val_l.append(p_val)

eda_data['shapiro_stats'] = stats_l
eda_data['p_val'] = p_val
```

```python
eda_data  = eda_data.reset_index()

x_vals = list(range(len(eda_data)))

fig = plt.subplots(figsize=(16,6)); sns.set_style('whitegrid')
sns.barplot( data = eda_data, x = 'col', y = 'percent_val')
plt.xticks(x_vals, eda_data['col'],  rotation='vertical')
plt.show()
```
