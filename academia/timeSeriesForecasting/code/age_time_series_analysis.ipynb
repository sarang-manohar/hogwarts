```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
rd = pd.read_csv(r'../data/Normalized_Age.csv',parse_dates=['Year'], index_col='Year'); rd.head()
```

```python
rd.plot(); plt.show()
```

```python
df_add_dcmps = seasonal_decompose(rd, model='additive')
df_add_dcmps.plot(); plt.show()
```

```python
rd['Age'] = rd['Age'] + 5 
```

```python
df_mult_dcmps = seasonal_decompose(rd, model='multiplicative')
df_mult_dcmps.plot(); plt.show()
```
