```python
tag_names = ["person", "bus", "car", "bicycle", "motorcycle"]
```

```python
# Import fiftyone
import fiftyone.zoo as fo

# Import data dataset 
def load_dataset(max_samples, seed):
    dataset_train = fo.load_zoo_dataset(
        "coco-2017",
        split="train",
        label_types=["detections"],
        classes=tag_names,
        max_samples=max_samples,
        seed=seed,
        dataset_name="oi-traffic-train" + str(seed),
    )
    return dataset_train
```

```python
dataset_train = load_dataset(max_samples=4000, seed=1)
```

```python
# set up connection to custom vision
credentials = ApiKeyCredentials(in_headers={"Training-key": creds['training_key']})
trainer = CustomVisionTrainingClient(creds['ENDPOINT'], credentials)
prediction_credentials = ApiKeyCredentials(in_headers={"Prediction-key": creds['prediction_key']})
predictor = CustomVisionPredictionClient(creds['ENDPOINT'], prediction_credentials)

#Create project
publish_iteration_name = "detect-traffic"

# Find the object detection domain
obj_detection_domain = next(domain for domain in trainer.get_domains() if domain.type == "ObjectDetection" and domain.name == "General")

# Create a new project
print ("Creating project...")
# Use uuid to avoid project name collisions.
project = trainer.create_project(str(uuid.uuid4()), domain_id=obj_detection_domain.id)

# Load the labels.json file
with open('/root/fiftyone/coco-2017/train/labels.json') as f:
    data = json.load(f)

# Create a dictionary to map image IDs to file names
image_id_to_file_name = {}
for image in data['images']:
    image_id_to_file_name[image['id']] = image['file_name']

# Create a dictionary to map category IDs to tag names
category_id_to_tag_name = {}
for category in data['categories']:
    category_id_to_tag_name[category['id']] = category['name']

# Define the Custom Vision project and tag names
project_id = project.id

# Get the Custom Vision project
project = trainer.get_project(project.id)

tags = []
for name in tag_names:
    tags.append(trainer.create_tag(project_id=project_id, name=name))

tag_dict = {}
for tag in trainer.get_tags(project_id):
    tag_dict[tag.name] = tag

# Get the Custom Vision tags
tags = {}
for tag_name in tag_names:
    tags[tag_name] = None
    for tag in trainer.get_tags(project_id):
        if tag.name == tag_name:
            tags[tag_name] = tag
            break
    if tags[tag_name] is None:
        tags[tag_name] = trainer.create_tag(project_id, tag_name)

# Create a list to store the image entries
image_entries = []

# Iterate over the images and annotations and create the image entries
for image in data['images']:
    file_name = image['file_name']
    image_path = f'/root/fiftyone/coco-2017/train/data/{file_name}'
    image_width = image['width']
    image_height = image['height']
    
    image_tags = []
    for annotation in data['annotations']:
        if annotation['image_id'] == image['id']:
            category_id = annotation['category_id']
            if category_id not in category_id_to_tag_name:
                continue
            tag_name = category_id_to_tag_name[category_id]
            if tag_name not in tag_names:
                continue
            
            bbox = annotation['bbox']
            left = bbox[0] / image_width
            top = bbox[1] / image_height
            width = bbox[2] / image_width
            height = bbox[3] / image_height
            
            region = Region(tag_id=tags[tag_name].id, left=left, top=top, width=width, height=height)
            image_tags.append(region)
    
    if len(image_tags) > 0:
        image_entry = ImageFileCreateEntry(name=file_name, contents=open(image_path, 'rb').read(), regions=image_tags)
        image_entries.append(image_entry)
    
    if len(image_entries) == 20:
        # Upload the images to Custom Vision
        upload_result = trainer.create_images_from_files(project_id, ImageFileCreateBatch(images=image_entries))
        image_entries = []

if len(image_entries) > 0:
    # Upload the remaining images to Custom Vision
    upload_result = trainer.create_images_from_files(project_id, ImageFileCreateBatch(images=image_entries))
```

```python
#Get project list
projects = trainer.get_projects()

project_list = []
for project in projects:
    project_list.append(project.id)

project_id = project_list[0]

#Train Model
new_iteration = trainer.train_project(project_id)

while (new_iteration.status != "Completed"):
    new_iteration = trainer.get_iteration(project_id, new_iteration.id)
    print ("Training status: " + new_iteration.status)
    time.sleep(1)
```
