### Checking outliers in the data

```python
import numpy as np
```

```python
outlier = []
for x in ['years_of_insurance_with_us', 'regular_checkup_lasy_year', 'visited_doctor_last_1_year', 'daily_avg_steps', 'age', 'avg_glucose_level', 'bmi', 'Year_last_admitted', 'weight', 'weight_change_in_last_one_year', 'fat_percentage']:
    q1, q3 = np.percentile(raw_data[x],[25,75])
    iqr = q3 - q1
    lower_bound = q1 -(1.5 * iqr) 
    upper_bound = q3 +(1.5 * iqr)
    for y in raw_data[x]:
        if y<lower_bound:
            outlier.append(raw_data[x])
        elif y>upper_bound:
            outlier.append(raw_data[x])
    #print(outlier)
    print(x,'\t',len(outlier)*100/25000,'%','\t',raw_data[x].median(),'\t',raw_data[x].min(),'\t',raw_data[x].max())
```

### Replacing outliers

```python
#a function to remove outliers from the data
def remove_outliers(df, col_name):
    q1, q3 = np.percentile(df[col_name],[25,75])
    iqr = q3 - q1
    lower_bound = q1 -(1.5 * iqr) 
    upper_bound = q3 +(1.5 * iqr)
    for x in df[col_name]:
        if x<lower_bound:
            df[col_name].replace(x, lower_bound, inplace=True)
        elif x>upper_bound:
            df[col_name].replace(x, upper_bound, inplace=True)
    return df
```

## K-Means clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

wss = []

for i in range(2,10):
    k_means = KMeans(n_clusters=i,random_state = 1)
    k_means.fit(df_scaled)
    wss.append(k_means.inertia_)

KMCluster = k_means.labels_

print(wss)

plt.plot(range(2,10), wss)
plt.axhline(y=489582.7244050991,color='y', linestyle = '--')
plt.axvline(x=3,color='y', linestyle = '--')
plt.title('The Elbow point graph')
plt.ylabel('Weighted Sums of Squares')
plt.xlabel('# clusters')
```

```python
s_c_comparison_test=pd.DataFrame()
min_Silhouette_sample_from_KMeans = []
Silhouette_score_from_KMeans = []
for n in np.arange(3,11,1):
    k_means = KMeans(n_clusters=n ,random_state = 1)
    k_means.fit(df_scaled)

    KMCluster = k_means.labels_
    min_Silhouette_sample_from_KMeans.append(silhouette_samples(df_scaled, KMCluster).min())
    Silhouette_score_from_KMeans.append(silhouette_score(df_scaled, KMCluster))
s_c_comparison_test['n_clusters'] = np.arange(3,11,1)
s_c_comparison_test['Silhouette_score'] = Silhouette_score_from_KMeans
s_c_comparison_test['min_Silhouette_sample'] = min_Silhouette_sample_from_KMeans
s_c_comparison_test = s_c_comparison_test.drop(index=5)
s_c_comparison_test
```

### Linear regression

```python
#Linear Regression
from sklearn.linear_model import LinearRegression

lnrReg = LinearRegression()
lnrReg.fit(X_train, y_train)  # fit the model
print(lnrReg.get_params())
y_train_pred = lnrReg.predict(X_train)  # make predictions
y_test_pred = lnrReg.predict(X_test)

print('R-square on train -', r2_score(y_train, y_train_pred))  # score the model on train data
print('R-square on test -', r2_score(y_test, y_test_pred))  # score the model on test data

print('Root Mean squared error on train -', np.sqrt(mean_squared_error(y_train, y_train_pred)))  # RMSE the model on train data
print('Root Mean squared error on test -', np.sqrt(mean_squared_error(y_test, y_test_pred)))  # RMSE the model on test data

print('Mean absolute percentage error on train -', mean_absolute_percentage_error(y_train, y_train_pred))  # MAPE the model on train data
print('Mean absolute percentage error on test -', mean_absolute_percentage_error(y_test, y_test_pred))  # MAPE the model on test data
```

```python
def base_model(mod):
    model = mod
    model.fit(X_train, y_train)  # fit the model
    #print(model.get_params())
    y_train_pred = model.predict(X_train)  # make predictions
    y_test_pred = model.predict(X_test)
    data = {
    'model': mod,
    'R2_train': r2_score(y_train, y_train_pred), 
    'R2_test': r2_score(y_test, y_test_pred), 
    'RMSE_train': np.sqrt(mean_squared_error(y_train, y_train_pred)), 
    'RMSE_test': np.sqrt(mean_squared_error(y_test, y_test_pred)), 
    'MAPE_train': mean_absolute_percentage_error(y_train, y_train_pred), 
    'MAPE_test': mean_absolute_percentage_error(y_test, y_test_pred)
    }
    return data
```
