```python
# import data
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
data = pd.read_csv('df_scaled.csv') # read data 
pd.set_option('display.max_columns', None) # display all columns

import warnings
warnings.filterwarnings('ignore')

data.head() # check the first 5 rows
```

```python
data.drop(['Unnamed: 0'], axis=1, inplace=True) # drop the first column
data.head() # check the first 5 rows
```

```python
raw_data = pd.read_csv('Data.csv') # read data
```

```python
# split data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, raw_data.iloc[:, -1], test_size=0.3, random_state=0)
```

### Linear regression

```python
#Linear Regression
from sklearn.linear_model import LinearRegression

lnrReg = LinearRegression()
lnrReg.fit(X_train, y_train)  # fit the model
print(lnrReg.score(X_train, y_train))  # score the model on train data
print(lnrReg.score(X_test, y_test))  # score the model on test data
```

```python
col_1 = list()
col_2 = list()

for idx, col_name in enumerate(X_train.columns):
    col_1.append(col_name)
    col_2.append(lnrReg.coef_[idx])

df_coef = pd.DataFrame.from_dict({'feature':col_1,'coef':col_2})

df_coef
```
