```python
import pandas as pd

# Load the data
df = pd.read_csv(r"../data/stock_data.csv")

# Calculate the 20-day moving average
df["ma"] = df["close"].rolling(window=20).mean()

df.sample()
```

```python
# Set the initial capital
initial_capital = 10000

# Create a column for the positions (either "long" or "short")
df["position"] = None

# Create a column for the PnL (profit and loss)
df["pnl"] = 0

# Create a column for the returns
df["return"] = 0

# Set the initial position to "long"
df.loc[0, "position"] = "long"

df.sample(10)
```

```python
# Iterate through the data and update the positions and PnL
for i in range(1, len(df)):
    # If the price crosses above the moving average, go long
    if df.loc[i, "close"] > df.loc[i, "ma"] and df.loc[i-1, "close"] <= df.loc[i-1, "ma"]:
        df.loc[i, "position"] = "long"
    # If the price crosses below the moving average, go short
    elif df.loc[i, "close"] < df.loc[i, "ma"] and df.loc[i-1, "close"] >= df.loc[i-1, "ma"]:
        df.loc[i, "position"] = "short"
    # If the position is "long", calculate the PnL and return
    elif df.loc[i, "position"] == "long":
        df.loc[i, "pnl"] = df.loc[i, "close"] - df.loc[i-1, "close"]
        df.loc[i, "return"] = df.loc[i, "pnl"] / df.loc[i-1, "close"]
    # If the position is "short", calculate the PnL and return
    elif df.loc[i, "position"] == "short":
        df.loc[i, "pnl"] = df.loc[i-1, "close"] - df.loc[i, "close"]
        df.loc[i, "return"] = df.loc[i, "pnl"] / df.loc[i-1, "close"]

df.sample(10)
```

```python
# Calculate the overall PnL and return
pnl = df["pnl"].sum()
returns = df["return"].sum()

# Print the results
print(f"PnL: {pnl}")
print(f"Returns: {returns}")
```
