```python
from NSEDownload import stocks
from datetime import datetime, date, timedelta

def convert_date_format(date_str):
  """
  Converts a date string from 'YYYY-MM-DD' format to 'd-m-yyyy' format.

  Args:
      date_str (str): The date string in 'YYYY-MM-DD' format.

  Returns:
      str: The converted date string in 'd-m-yyyy' format.
  """
  try:
      # Parse the date string
      date_obj = datetime.strptime(date_str, "%Y-%m-%d")
      # Format the date object with the desired format (corrected specifier)
      new_date_str = date_obj.strftime("%d-%m-%YYYY")  # Use 'YYYY' for four-digit year
      return new_date_str
  except ValueError:
      print(f"Invalid date format for: {date_str}")
      return None

today = str(date.today())



# Gets data without adjustment for events
df = stocks.get_data(stock_symbol="ITBEES", start_date='1-1-2024', end_date=today)
df
```

```python
import pandas as pd
import yfinance as yf  # Use yfinance instead of googlefinance
from datetime import date, timedelta
import plotly.graph_objects as go

# Define list of securities and benchmark
sectors = ["ITBEES.NS", "PSUBNKBEES.NS", "CONSUMBEES.NS", "INFRABEES.NS", "HNGSNGBEES.NS",
           "AUTOBEES.NS", "GILT5YBEES.NS", "LTGILTBEES.NS", "BANKBEES.NS", "LIQUIDBEES.NS",
           "PHARMABEES.NS", "GOLDBEES.NS", "SILVERBEES.NS"]

benchmark = "NIFTYBEES.NS"

# Get today's date
today = date.today()

# Calculate start date (5 years ago)
five_years_ago = today - timedelta(days=365 * 5)

# Empty dictionary to store closing prices
closing_prices = {}

historical_data = pd.DataFrame(columns=['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'Script'])

# Loop through each security and benchmark
for security in sectors + [benchmark]:
  # Download historical data using yfinance
  data = yf.download(security, start=five_years_ago, end=today)
  data['Script'] = security
  historical_data = pd.concat([historical_data,data])

  # Check if data is downloaded successfully (avoid errors for missing data)
  if data.empty:
      print(f"Data not available for {security}")
      continue
```
